# Time Resampling

```python
df.head()
```

```python
df.resample(rule = 'A').min()
```

```python
df.resample(rule = 'A').max()
```

```python
df.resample(rule = 'A').max()['Open'].plot()
```

```python
df.resample(rule = 'A').max()['High'].plot()
```

```python
df.resample(rule = 'A').max()['Low'].plot()
```

```python
df.resample(rule = 'A').max()['Volume'].plot()
```

```python
# Frequency Quaterly
df.resample(rule = 'QS').max()
```

```python
# Frequency  Quaterly
df.resample(rule = 'QS').max()["High"].plot()
```

```python
# Frequency Business Year
df.resample(rule = 'BA').max()
```

```python
df.resample(rule = 'BA').max()["Volume"].plot()
```

```python
df.resample(rule = 'BA').max()['Open'].plot()
```

```python
# Frequency  Business Quater
df.resample(rule = 'BQS').max()
```

```python
df.resample(rule = 'BQS').max()['Open'].plot()
```

```python
# Weekday
df.resample(rule = 'W').max()
```

```python
df.resample(rule = 'W').max()['Open'].plot()
```

```python
df['Open'].resample(rule = 'BA').mean().plot(kind = 'bar')
```
