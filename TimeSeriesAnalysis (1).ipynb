```python
# Cummulative Moving Average ..

## Expanding is a function used for Commulative Moving Average

df['Open'].expanding().mean().plot(figsize = (10,5))
# Average is Increasing Which means Stock Price is also increasing ..
```

```python
# Exponentail Moving Average ...
df['EMA_Open_0.1'] = df['Open'].ewm(alpha = 0.1 , adjust = False).mean()
```

```python
df[['Open','EMA_Open_0.1']].plot(xlim = ['2020-01-01','2021-01-01'],figsize = (15,6))
```

```python
df['EMA_Open_0.3'] = df['Open'].ewm(alpha = 0.3 , adjust = False).mean()
```

```python
df[['Open','EMA_Open_0.1','EMA_Open_0.3']].plot(xlim = ['2020-01-01','2021-01-01'],figsize = (16,9))
```

```python
df['EMA_5days'] = df['Open'].ewm(span = 5).mean()
```

```python
df[['Open','EMA_Open_0.1','EMA_Open_0.3','EMA_5days']].plot(xlim = ['2020-01-01','2021-01-01'],figsize = (16,9))
```
